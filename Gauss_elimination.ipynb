```python
matrix = [[0,0,1,1],
          [2,2,2,0],
          [2,2,1,-1],]
mat = matrix
```

```python
def print_mat(mat): #matrix 출력
    for r in range(len(mat)):
        for c in range(len(mat[r])):
            print(mat[r][c], end = '\t')
        print('\n')
        
print_mat(mat)
```

```python
matrix = [[0,0,1,1],
          [2,2,2,0],
          [2,2,1,-1],]



mat = matrix.copy()
#find row for pivot
def find_pivot(mat, col): #col은 기준 column
    pivot_r = col #현재 pivot row
    all_zero = 0
    find_p = col
    if mat[pivot_r][col] == 0: #현재 진행해야하는 행, 열의 값이 0이면, 해당 열 중 0이 아닌 값을 찾는다.
        find_p = pivot_r    
        while mat[find_p][pivot_r] == 0:
            if find_p == len(mat)-1: #해당 pivot의 모든 열 값이 0, 마지막 행인 경우
                all_zero = 1 #0이 아닌 pivot이 있다. 
                break
            find_p+=1
            
    if all_zero == 0: #0이 아닌 pivot 존재하면
        mat[find_p], mat[pivot_r] = mat[pivot_r], mat[find_p] #pivot행 교환
        print('row changed')
    return mat

mat = find_pivot(mat, 0)
print_mat(mat)
print()

#피봇 행 곱한 후 덧셈
def mul_add(mat, col):
    pivot_r = col
    factor = 1
    #우선 피봇 행을 자기 자신의 피봇 열로 나누어 준다.
    if mat[pivot_r][col] != 0: #0으로 나눈 값이 들어갈 수 없으므로 이 경우는 따로 처리해 준다.
        factor = mat[pivot_r][col]
#         for c in range(len(mat[pivot_r])):
#             mat[pivot_r][c] = mat[pivot_r][c]/factor

    if col < len(mat) : #전체 행의 수보다 적어야 한다.      
        for r in range(col+1, len(mat)): 
            cur_pivot_col = mat[r][col]
            for c in range(len(mat[r])):
                mat[r][c] -= mat[pivot_r][c] * (cur_pivot_col/factor) 
                
    return mat
mat = mul_add(mat, 0)

print_mat(mat)
print('\n\nstart\n\n')


#back substraction
def back_sub(mat):
    #number of unknowns
    num_unk = len(mat[0])-1
    sol_list = [] #해가 담길 리스트
    while num_unk>0: #back substraction
        pivot_r = num_unk-1
        if mat[pivot_r][num_unk-1] == 0: #0이면 무수히 많은 해
            sol_list = ['inf']
            break
        factor = mat[pivot_r][num_unk]
        for c in range(len(mat[pivot_r])):
            mat[pivot_r][c] = mat[pivot_r][c]/factor #행을 자신의factor로 나누기
            
        
        if len(sol_list)==0: #가장 마지막 행, 아직 해가 리스트에 안 담겨 있다.
            sol_list.append(mat[pivot_r][-1]) #가장 마지막 열이 b의 값
            print(3)
        else:
            temp_sol = 0
            temp_sol = mat[pivot_r][-1]
            for i, sol in enumerate(sol_list):
                temp_sol -= sol * mat[pivot_r][num_unk-i]
            sol_list.append(temp_sol)
        num_unk -= 1
    sol_list.reverse()
    return sol_list
        
        
##################################################################    
    
    
def gauss_elimination(mat):
    row = len(mat)
    col = len(mat[0])-1

    for c in range(col):
        mat = find_pivot(mat, c)
        mat = mul_add(mat, c)

    print_mat(mat)


    sol_list = back_sub(mat)
    return sol_list



matrix = [[0,0,1,1],
          [2,2,2,0],
          [2,2,1,-1],]
mat = matrix.copy()

print_mat(mat)
print('-----------------------------')
#전체 처리, mat은 augmented matrix임을 고려한다. n x m+1
row = len(mat)
col = len(mat[0])-1

for c in range(col):
    print(c+1, ') ')
    mat = find_pivot(mat, c)
    print_mat(mat)
    print()
    mat = mul_add(mat, c)
    print_mat(mat)
    print()
    
print_mat(mat)


sol_list = back_sub(mat)
print(sol_list)
    
    
```
